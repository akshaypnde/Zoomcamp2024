```python
engine = create_engine("postgresql://root:root@localhost:5432/ny_taxi")
```

```python
schema_p = pd.io.sql.get_schema(df_ny_taxi_2023, name="yellow_taxi_data", con=engine)
schema_p
```

```python
print(schema_p)
```

```python
pq_file = pyarrow.parquet.ParquetFile("yellow_tripdata_2023-01.parquet")

batches = pq_file.iter_batches()
batches

#df_iter = pd.read_parquet("yellow_tripdata_2023-01.parquet", iterator=True, chunksize=100000) 
```

```python
df = next(batches).to_pandas()
```

```python
df.head(10)
```

```python
df.to_sql(name="yellow_taxi_data", con=engine, if_exists="replace")
```

```python
df.to_sql(name="yellow_taxi_data", con=engine, if_exists="append")
```

```python
from time import time
```

```python
while True:
    start_time = time()
    
    try:
        df = next(batches).to_pandas()
    except Exception:
        break
        
    pd.to_datetime(df.tpep_pickup_datetime)
    pd.to_datetime(df.tpep_dropoff_datetime)
    
    df.to_sql(name="yellow_taxi_data", con=engine, if_exists="append")

    end_time = time()
    print("Inserting chunk.. took % .3f seconds " % (end_time - start_time))
```

```python
engine.connect()
```
